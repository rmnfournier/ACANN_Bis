```python
import numpy as np
import pandas as pd 

good_parameters = pd.read_csv("parameters_training.csv",header=None)
wrong_parameters=pd.read_csv("err_parameters_training.csv",header=None)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

plt.figure(figsize=(15,6))
for i in np.arange(0,7):
    plt.subplot(2,7,i+1)
    sns.distplot(good_parameters[i],bins=100);
    plt.subplot(2,7,i+8)
    sns.distplot(wrong_parameters[i],bins=100);

```

```python
import scipy.stats as sts
plt.figure(figsize=(5,5))
sns.heatmap(good_parameters.corr())
plt.figure(figsize=(5,5))
sns.heatmap(wrong_parameters.corr())
```
